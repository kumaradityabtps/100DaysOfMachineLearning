```python
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sh vs V Kohli career Comparision')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')

plt.title('Rohit Sh vs V Kohli career Comparision')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',linewidth=8)

plt.title('Rohit Sh vs V Kohli career Comparision')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',linewidth=8)

plt.title('Rohit Sh vs V Kohli career Comparision')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',linewidth=8,label="Rohit")

plt.title('Rohit Sh vs V Kohli career Comparision')
plt.xlabel('season')
plt.ylabel('runs')

plt.legend()
```
